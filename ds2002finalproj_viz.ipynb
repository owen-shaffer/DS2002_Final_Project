```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


url = 'https://raw.githubusercontent.com/owen-shaffer/DS2002_Final_Project/main/Crime_Data.csv'

Crime_cville = pd.read_csv(url)

Crime_cville.head(20)
```

```python
crimes = Crime_cville.Offense.unique()
crimes
```

```python
Crime_cville['DateReported'] = pd.to_datetime(Crime_cville['DateReported'])

Cville_2023 = Crime_cville[Crime_cville['DateReported'].dt.year == 2023]

offense_counts_2023 = Cville_2023['Offense'].value_counts()

total_crimes_2023 = len(Cville_2023)
print(f"Total number of crimes in 2023: {total_crimes_2023}")
```

```python
top_20_offenses = offense_counts_2023.head(20)

plt.figure(figsize=(12, 6))
top_20_offenses.plot(kind='bar')
plt.title('Total Number of Each Offense')
plt.xlabel('Offense')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
larceny_crimes = Cville_2023[Cville_2023['Offense'].str.contains('Larceny', case=False, na=False)]
df = pd.DataFrame(larceny_crimes)
df.head()
```

```python

assault_crimes = Cville_2023[Cville_2023['Offense'].str.contains('Assault', case=False, na=False)]

assault_crimes['Month'] = assault_crimes['DateReported'].dt.month
monthly_assault_counts = assault_crimes.groupby('Month')['Offense'].count()

group1_months = [1, 5, 6, 7, 8]
group2_months = [month for month in range(1, 13) if month not in group1_months]

avg_group1 = monthly_assault_counts[monthly_assault_counts.index.isin(group1_months)].mean()
avg_group2 = monthly_assault_counts[monthly_assault_counts.index.isin(group2_months)].mean()

print(f"Average assault crimes per month without students (Months 1, 5, 6, 7, 8): {avg_group1}")
print(f"Average assault crimes per month with students: {avg_group2}")
```

```python
# Graph the number of crime reported in 2023 by hour

import pandas as pd
import matplotlib.pyplot as plt

# Charlottesville
Cville_2023['DateReported'] = pd.to_datetime(Cville_2023['DateReported'])
Cville_2023['Hour'] = Cville_2023['DateReported'].dt.hour
crime_by_hour_cville = Cville_2023.groupby('Hour')['Offense'].count()

plt.figure(figsize=(10, 6))
plt.plot(crime_by_hour_cville.index, crime_by_hour_cville.values, marker='o', color="purple")
plt.title('Number of Crimes Reported by Hour in Charlottesville (2023)')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Crimes')
plt.xticks(range(24))
plt.grid(True)
plt.show()


# Ann Arbor
AnnArbor_2023['DATE'] = pd.to_datetime(AnnArbor_2023['DATE'])
AnnArbor_2023['Hour'] = AnnArbor_2023['DATE'].dt.hour
crime_by_hour_annarbor = AnnArbor_2023.groupby('Hour')['CATEGORY'].count()

plt.figure(figsize=(10, 6))
plt.plot(crime_by_hour_annarbor.index, crime_by_hour_annarbor.values, marker='o', color='blue')
plt.title('Number of Crimes Reported by Hour in Ann Arbor (2023)')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Crimes')
plt.xticks(range(24))
plt.grid(True)
plt.show()
```

```python
#Setting up the key
import os
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = "/content/GoogleKey.json"

from google.cloud import storage
```

```python
bucket_name = "ds_final_project"
cville_file_name = "/content/Cville_2023.csv"
cville_destination_blob_name = "data/Cville_2023.csv"

Ann_Arbor_file_name = "/content/AnnArbor_2023.csv"
annarbor_destination_blob_name = "data/AnnArbor_2023.csv"


def upload_to_gcs(bucket_name, source_file_name, destination_blob_name):
    client = storage.Client()
    bucket = client.bucket(bucket_name)
    blob = bucket.blob(destination_blob_name)
    blob.upload_from_filename(source_file_name)

upload_to_gcs(bucket_name, cville_file_name, cville_destination_blob_name)
upload_to_gcs(bucket_name, Ann_Arbor_file_name, annarbor_destination_blob_name)
```
